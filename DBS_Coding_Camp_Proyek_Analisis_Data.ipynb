### `customer_zip_code_prefix`

### `customer_city`
- Nama kota tempat pelanggan berada.
- Kolom ini dapat digunakan untuk:

### `customer_state`
- Kode **negara bagian** di Brasil (contoh: `SP` untuk São Paulo).

### Insight per Kolom – Data Order Items (`order_items`)

Tabel `order_items` berisi detail tiap produk dalam setiap pesanan pelanggan. Satu `order_id` bisa memiliki lebih dari satu `order_item_id` jika pelanggan membeli lebih dari satu produk dalam satu transaksi. Tabel ini penting untuk analisis penjualan, performa produk, dan biaya logistik.

### `order_id`
- ID unik pesanan pelanggan.

---

### `order_item_id`
- Merupakan **penanda urutan** item dalam satu `order_id`.
- Biasanya dimulai dari 1 (untuk item pertama), dan bertambah jika terdapat lebih dari satu produk dalam satu pesanan.
- Penting untuk menghitung jumlah item per pesanan (misal: total produk dibeli per order).

---

### `product_id`
- ID unik dari produk yang dibeli.
- Digunakan untuk menggabungkan data ke tabel `products`.
- Dapat digunakan untuk menghitung:
  - Produk yang paling sering dibeli
  - Produk dengan nilai penjualan tertinggi

---

### `seller_id`
- ID penjual yang menjual produk tersebut.
- Setiap produk dalam satu pesanan bisa berasal dari **penjual berbeda**.

  - Menganalisis performa penjual

---

### `shipping_limit_date`
- Batas waktu pengiriman dari penjual ke kurir.

  - Menghitung **lead time seller**

### `price`

  - Produk dengan harga tertinggi/rendah
  - Produk dengan total penjualan tertinggi

### `freight_value`
- Biaya ongkir yang dibayarkan oleh pelanggan untuk item tersebut.

  - Menghitung **total biaya logistik**

Tabel `order_payments` merekam informasi metode pembayaran, jumlah cicilan, serta total nilai yang dibayarkan pelanggan per pesanan. Satu pesanan (`order_id`) dapat memiliki lebih dari satu metode pembayaran jika pelanggan membayar sebagian dengan metode berbeda (misal: sebagian dengan voucher, sebagian dengan kartu kredit).

### `order_id`
- ID unik untuk setiap pesanan pelanggan.

### `payment_sequential`
- Menunjukkan urutan metode pembayaran dalam satu pesanan (1, 2, dst.).
- Biasanya bernilai `1`, namun bisa lebih jika pembayaran dilakukan dengan lebih dari satu metode.
- Cocok untuk mendeteksi **multi-payment strategy** (misal: kombinasi kartu kredit dan voucher).

---

### `payment_type`
- Jenis metode pembayaran yang digunakan, seperti:
  - `credit_card`
  - `boleto` (mirip giro/kertas bayar)

  - `debit_card`

### `payment_installments`
- Jumlah cicilan yang diambil pelanggan untuk menyelesaikan pembayaran.

  - Mengetahui **perilaku keuangan pelanggan**
  - Mengukur **ketergantungan terhadap sistem cicilan**

---

### `payment_value`
- Total jumlah uang yang dibayarkan oleh pelanggan untuk pesanan tersebut.
- Sudah termasuk ongkir (`freight_value`) dan harga produk (`price`) dalam satuan BRL (Brazilian Real).

  - Menghitung **total pendapatan per pesanan**
  - Analisis **rata-rata nilai pembelian (Average Order Value)**

---

> Tabel `order_payments` sangat penting untuk analisis keuangan pelanggan, strategi promosi metode pembayaran, dan pengelolaan arus kas berdasarkan preferensi cicilan.

## 3.2 Transformasi Tipe Data

```python
# 3.2 Transformasi Tipe Data

# Konversi kolom tanggal ke tipe datetime
datetime_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'review_creation_date',
    'review_answer_timestamp'
]

for col in datetime_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Konversi kolom kategorikal ke tipe category
category_cols = [
    'order_status',
    'payment_type',
    'customer_state',
    'customer_city',
    'seller_state',
    'seller_city',
    'product_category_name',
    'product_category_name_english'
]

for col in category_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

# Verifikasi hasil transformasi
print("Info dataset setelah transformasi tipe data:")
df.info()
```

```python
# 3.3 Feature Engineering

# 1. Bulan & Hari Pembelian (untuk analisis musiman - Q2)
df['purchase_month'] = df['order_purchase_timestamp'].dt.month
df['purchase_dayofweek'] = df['order_purchase_timestamp'].dt.dayofweek  # 0 = Senin, 6 = Minggu

# 2. Durasi pengiriman ke pelanggan (untuk analisis performa logistik / Q3)
df['delivery_duration'] = (
    df['order_delivered_customer_date'] - df['order_purchase_timestamp']
).dt.days

# 3. Estimasi delay: apakah pengiriman terlambat?
df['delivery_delay'] = (
    df['order_delivered_customer_date'] > df['order_estimated_delivery_date']
)

# 4. Total nilai transaksi: harga produk + ongkir (untuk RFM - Monetary)
df['total_payment'] = df['price'] + df['freight_value']

# 5. Tahun transaksi (opsional untuk segmentasi tahunan)
df['purchase_year'] = df['order_purchase_timestamp'].dt.year

# Verifikasi hasil
df[['purchase_month', 'purchase_dayofweek', 'delivery_duration', 'delivery_delay', 'total_payment']].head()
```

## Analisis Lanjutan RFM
Pertanyaan 3: Bagaimana Cara Mengelompokkan Pelanggan Berdasarkan Pola Pembelian Mereka? (RFM Analysis)

```python
# RFM Analysis – Segmentasi Pelanggan berdasarkan Recency, Frequency, dan Monetary
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Hitung nilai RFM
rfm_df = df.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (df['order_purchase_timestamp'].max() - x.max()).days,
    'order_id': 'nunique',
    'total_payment': 'sum'
}).reset_index()

# 2. Rename kolom hasil agregasi
rfm_df.columns = ['customer_id', 'Recency', 'Frequency', 'Monetary']

# 3. Skor kuantil RFM (semakin bagus, skor semakin tinggi)
rfm_df['R_score'] = pd.qcut(rfm_df['Recency'], q=4, labels=[4, 3, 2, 1])  # Recency: makin kecil makin bagus
rfm_df['F_score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), q=4, labels=[1, 2, 3, 4])  # Frequency: makin besar makin bagus
rfm_df['M_score'] = pd.qcut(rfm_df['Monetary'], q=4, labels=[1, 2, 3, 4])  # Monetary: makin besar makin bagus

# 4. Gabungkan skor menjadi RFM_Score string
rfm_df['RFM_Score'] = (
    rfm_df['R_score'].astype(str) +
    rfm_df['F_score'].astype(str) +
    rfm_df['M_score'].astype(str)
)

# 5. Segmentasi berdasarkan skor gabungan
def segment_customer(score):
    score_int = int(score)
    if score_int >= 344:
        return 'High Value'
    elif score_int >= 222:
        return 'Mid Value'
    else:
        return 'Low Value'

rfm_df['Segment'] = rfm_df['RFM_Score'].apply(segment_customer)

# 6. Visualisasi jumlah pelanggan per segmen
plt.figure(figsize=(8, 5))
sns.countplot(
    data=rfm_df,
    x='Segment',
    order=['Low Value', 'Mid Value', 'High Value'],
    palette='viridis'
)
plt.title('Segmentasi Pelanggan Berdasarkan RFM')
plt.xlabel('Segmentasi')
plt.ylabel('Jumlah Pelanggan')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

- **Mid Value Customers**:

### Kesimpulan:
Pelanggan e-commerce terbagi cukup seimbang, namun dominasi **Mid Value** menandakan peluang besar untuk meningkatkan loyalitas dan konversi.  
**Strategi diferensiasi per segmen RFM** dapat membantu meningkatkan lifetime value pelanggan secara signifikan.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])

# Hitung delivery_duration (dalam hari)
df['delivery_duration'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

# Filter data yang valid (tanpa null dan delivery_duration positif)
valid_df = df[['delivery_duration', 'review_score']].dropna()
valid_df = valid_df[valid_df['delivery_duration'] >= 0]

# Korelasi Pearson
corr_value, _ = pearsonr(valid_df['delivery_duration'], valid_df['review_score'])

# Visualisasi scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=valid_df, x='delivery_duration', y='review_score', alpha=0.4)
sns.regplot(data=valid_df, x='delivery_duration', y='review_score', scatter=False, color='red', line_kws={"linewidth": 2})
plt.title(f'Korelasi Durasi Pengiriman vs Review Score (ρ = {corr_value:.2f})')
plt.xlabel('Durasi Pengiriman (hari)')
plt.ylabel('Skor Review')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

## Insight Korelasi: Durasi Pengiriman vs Skor Review

### Temuan Visualisasi:

- Terdapat **korelasi negatif moderat** antara `delivery_duration` dan `review_score` dengan nilai **ρ = -0.30**.
- Artinya, **semakin lama pengiriman**, cenderung semakin **rendah skor review** yang diberikan oleh pelanggan.
- Hal ini juga terlihat dari **regresi linear menurun** (garis merah), di mana banyak skor 5 berada pada pengiriman < 20 hari, sedangkan skor 1 tersebar luas di rentang >30 hari.

---

### Insight Kunci:

- **Review tertinggi (skor 5)** sangat banyak muncul di pengiriman yang cepat (<10 hari).
- Skor rendah (1–2) lebih sering muncul pada pengiriman yang **melewati waktu normal**, bahkan hingga ratusan hari (outlier).
- Ini mengindikasikan bahwa **durasi pengiriman adalah salah satu faktor penting** yang memengaruhi kepuasan pelanggan.

---

### Implikasi Bisnis:

- **Performa logistik** sangat memengaruhi persepsi pelanggan terhadap layanan.
- Perlu dilakukan:
  - Optimasi durasi kirim untuk menjaga **nilai brand dan retensi pelanggan**
  - Pemantauan keterlambatan sebagai **indikator risiko churn**
  - Segmentasi review rendah berdasarkan durasi kirim ekstrem untuk audit vendor atau kurir

### Kesimpulan:
Terdapat **korelasi negatif yang cukup kuat** antara lama pengiriman dan skor review.  
Semakin cepat produk dikirim, semakin besar kemungkinan pelanggan memberi penilaian yang tinggi.  
E-commerce perlu menjadikan **kecepatan pengiriman** sebagai prioritas untuk meningkatkan **kepuasan dan loyalitas pelanggan**.

Produk dari kategori **`bed_bath_table`**, **`health_beauty`**, dan **`sports_leisure`** merupakan kategori yang paling sering dibeli pelanggan.

### Pertanyaan 2: Kapan Transaksi Paling Banyak Terjadi Sepanjang Tahun?
Puncak transaksi terjadi pada **bulan Mei hingga Agustus**, dengan Agustus sebagai bulan tertinggi.  
Secara mingguan, **awal pekan (Senin dan Selasa)** mencatat jumlah transaksi tertinggi.  
*Implikasi bisnis*: Promosi ideal dilakukan pada kuartal 2–3 dan di awal pekan untuk memaksimalkan konversi pembelian.

### Pertanyaan 3: Bagaimana Cara Mengelompokkan Pelanggan Berdasarkan Pola Pembelian Mereka? (RFM Analysis)
Pelanggan dikelompokkan menggunakan metode **RFM (Recency, Frequency, Monetary)** menjadi:
- **High Value**: pelanggan loyal dan aktif
- **Mid Value**: pelanggan reguler yang potensial ditingkatkan loyalitasnya
- **Low Value**: pelanggan tidak aktif yang berisiko churn

Sebagian besar pelanggan termasuk dalam segmen **Mid Value**, menandakan peluang besar untuk ditingkatkan menjadi High Value melalui strategi loyalitas, upsell, dan re-engagement.

*Implikasi bisnis*: Segmentasi ini memungkinkan pendekatan pemasaran yang lebih tepat sasaran dan efisien.
